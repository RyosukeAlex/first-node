```python
# API keyの設定（環境変数）
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
# Chat Completions API Call

from openai import OpenAI

client = OpenAI()

responce = client.chat.completions.create(
    model = 'gpt-4o-mini',
    messages=[
        {'role': 'system', 'content': 'You are a helphul assistant.'},
        {'role': 'user', 'content': 'こんにちは。私はりけといいます。'},
    ]
)

print(responce.to_json(indent=2))
```

```python
# 過去のやり取りを参照させる方法
responce = client.chat.completions.create(
    model='gpt-4o-mini',
    messages=[
        {'role': 'system', 'content': 'You are hulpfull assistant'},
        {'role': 'user', 'content': 'こんにちは。私はりけといいます。'},
        {'role': 'assistant', 'content': 'こんにちは、りけさん！お会いできてうれしいです。何かお手伝いできることがありますか？'},
        {'role': 'user', 'content': '私の名前は覚えていますか？'},
    ],
)

print(responce.to_json(indent=2))
```

```python
# ストリーミングレスポンス 
responce = client.chat.completions.create(
    model = 'gpt-4o-mini',
    messages=[
        {'role': 'system', 'content': 'You are helpfull guy.'},
        {'role': 'user', 'content': '私はNode On株式会社の代表です。'}
    ],
    stream=True
)

for chunk in responce:
    content = chunk.choices[0].delta.content
    if content is not None:
        print(content, end='', flush=True)
```

```python
# JSONモード
responce = client.chat.completions.create(
    model='gpt-4o-mini',
    messages=[
        {
            'role':'system',
            'content':'人物一覧を次のJSON形式で出力してください。\n{"people": ["aaa","bbb"]}',
        },
        {
            'role':'user',
            'content':'NodeOnのCEOは谷口、COOは福盛です。'
        },
    ],
    response_format={'type':'json_object'},
)

print(responce.choices[0].message.content)
```
